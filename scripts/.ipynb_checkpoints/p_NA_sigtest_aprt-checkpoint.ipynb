THIS PYTHON SCRIPT READS VARIOUS MODEL OUTPUTS (here: monthly resolved ECHAM5 atmosphere outputs at T31 resolution).
BASED ON CDO COMMANDS THAT ARE CALLED FROM WITHIN THE SCRIPT, THE MODEL OUTPUTS ARE PREPROCESSED (time average for annual, DJF, and JJA; anomaly) AND ANOMALIES WITH RESPECT TO A REFERENCE PERIOD ARE PLOTTED ON A GEOGRAPHIC MAP; TO GET AN IDEA OF UNCERTAINTY OVER TIME. THE SIGNIFICANCE OF ANOMALIES IS COMPUTE BASED ON A T-TEST AND NON-SIGNIFICANT ANOMALIES ARE IDENTIFIED VIA HATCHING.

VARIOUS SIMULATIONS ARE PLOTTED VIA A LOOP, THE SET OF SIMULATIONS TO BE PLOTTED CAN BE MODIFIED VIA  ADJUSTING / APPENDING TO A PYTHON LIST WHERE INPUT PATHS AND FILE NAMES ARE COLLECTED

```python
# import necessary packages
import matplotlib.pyplot as plt
import numpy as np
from netCDF4 import Dataset
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
import matplotlib.colors as colors
import os as os
import scipy.stats
import matplotlib as mpl
```

```python
#define auxiliary functions that are employed for significance testing
#functions and procedure taken over from Paul Gierz

os.chdir('/work/ollie/lsijbran/climate-feedbacks-proglacial-lakes/z_jupyternote/400y_study/EchamVars/significance/NorthAmerica/') #LS adjust

def estimated_autocorrelation(x):
    n=len(x)
    variance=x.var()
    x=x-x.mean()
    r=np.correlate(x, x, mode='full')[-n:]
    result=r/(variance*(np.arange(n,0,-1)))
    return result

def compute_significance(experiment,reference,lons,lats,mask_cutoff,verbose=False):
    sig_mask=np.zeros((len(lats),len(lons)))
    sig_mask[:]=np.nan
    n_time=experiment.shape[0]
    for i in range(len(lons)):
        for j in range(len(lats)):
            xx=experiment[:, j, i]
            yy=reference[:, j, i]
            autocorrx=max(estimated_autocorrelation(xx)[1], 0)
            autocorry=max(estimated_autocorrelation(yy)[1], 0)
            eff_dof1=n_time * (1 - autocorrx) / (1 + autocorrx)
            eff_dof2=n_time * (1 - autocorry) / (1 + autocorry)
            eff_dof_comb = min(eff_dof1, eff_dof2)
            cutoff=scipy.stats.t.ppf(mask_cutoff, eff_dof_comb)
            t_test = abs(xx.mean()-yy.mean())/((xx.var()/n_time+yy.var()/n_time)**0.5)
            if t_test<cutoff:
                sig_mask[j,i]=1
            else:
                sig_mask[j,i]=0
            if verbose:
                print(cutoff, t_test)
                print(t_test < cutoff)
    return sig_mask



#definition of variable name that will be plotted (here: temp2)
variable="aprt" #LS adjust



#%%definition of parameter that controls plotting of the colorbar
plot_cbar=True



#%%definition of a proper (i.e. thinner than standard) linewidth for hatchs
mpl.rcParams['hatch.linewidth']=0.1



#%%definition of data files to plot
experiment_file=[]
experiment_file.append('/work/ollie/lsijbran/climate-feedbacks-proglacial-lakes/z_jupyternote/400y_study/EchamVars/significance/NorthAmerica/plake_2250-2349_aprt_mm.nc') #('Eoi400_4th_try_echam5_main_mm_2650-2749_temp2.nc') LS adjust
reference_file='/work/ollie/lsijbran/climate-feedbacks-proglacial-lakes/z_jupyternote/400y_study/EchamVars/significance/NorthAmerica/alakeGLAC_2250-2349_aprt_mm.nc'  #'E280_280ppm_echam5_main_mm_2650-2749_temp2.nc' LS adjust
```

```python
#%%loop over the simulations and seasons and create anomaly maps for each
for exp in experiment_file:
    print("")
    print("")
    print("processing simulation " + exp + " ...")
    #loop processing various time averages, details of computations to be defined below
    for j in "DJF","JJA","ANN":
        print("")
        print("processing season " + j + " ...")
        #%%define various file names
        #graphics output file name (adjusted to the seasonal mean being plotted)
        outfile=exp.replace('2250-2349_aprt_mm.nc','glac_'+j+'_NA_'+variable+'.png')#LS adjust: do not forget to adjust the filename to plotted variable and reference period
        #output file for reference timeseries
        outfile_ref_timeser="reference_" + j + ".nc"
        #output file for experiment timeseries
        outfile_exp_timeser=exp.replace(".nc","") + "_" + j + ".nc"
        #output file for reference seasonal mean
        outfile_ref_seasmean=outfile_ref_timeser.replace(".nc", "_yseasmean.nc")
        #output file for experiment seasonal mean
        outfile_exp_seasmean=outfile_exp_timeser.replace(".nc", "_yseasmean.nc")
        
        #%%generate seasonal time series for both reference and simulation
        if j=="DJF":
            mon_str="12,1,2"
        if j=="JJA":
            mon_str="6,7,8"
        if j=="ANN":
            mon_str="1/12"
        #for the following commands, do not forget to adjust the number of timesteps (here, the input data is assumed to have 1200 monhtly means)
        cmd="cdo seasmean -selmon," + mon_str + " -seltimestep,3/1199 " + reference_file + " " + outfile_ref_timeser #we cut off the first two and last month to avoid incomplete winter seasons) #LS (1199 statt 359) LS adjust
        print("executing command: " + cmd)
        os.system(cmd)
        cmd="cdo seasmean -selmon," + mon_str + " -seltimestep,3/1199 " + exp + " " + outfile_exp_timeser #LS (vorher 1199 statt 359) LS adjust
        print("executing command: " + cmd)
        os.system(cmd)
    
        #%%generate multiyear seasonal average for both reference and simulation
        cmd="cdo timmean " + outfile_ref_timeser + " " + outfile_ref_seasmean
        print("executing command: " + cmd)
        os.system(cmd)
        cmd="cdo timmean " + outfile_exp_timeser + " " + outfile_exp_seasmean
        print("executing command: " + cmd)
        os.system(cmd)
        
                #%%read seasonal mean NetCDF data for reference and experiment
        print("read data from file " + outfile_ref_seasmean)
        data=Dataset(outfile_ref_seasmean)
        ref_seasmean_data=data.variables[variable][:].squeeze()
        ref_model_lon=data.variables['lon'][:]
        ref_model_lat=data.variables['lat'][:]
        print("read data from file " + outfile_exp_seasmean)
        data=Dataset(outfile_exp_seasmean)
        exp_seasmean_data=data.variables[variable][:].squeeze()
        
        #%%read seasonal timeseries NetCDF data for reference and experiment
        print("read data from file " + outfile_ref_timeser)
        data=Dataset(outfile_ref_timeser)
        ref_seas_data=data.variables[variable][:].squeeze()
        ref_model_timeser_lon=data.variables['lon'][:]
        print("read data from file " + outfile_exp_timeser)
        data=Dataset(outfile_exp_timeser)
        exp_seas_data=data.variables[variable][:].squeeze()
        exp_model_timeser_lon=data.variables['lon'][:]
    
        #%%compute seasonal mean anomaly
        anomaly=exp_seasmean_data-ref_seasmean_data
        
        #%%close the data gap in the modelled seasonal mean anomaly by adding cyclic point
        anomaly, ref_model_lon=add_cyclic_point(anomaly, coord=ref_model_lon)
        
        #%%close data gap in the modelled seasonal time series by adding cyclic point
        ref_seas_data, ref_model_timeser_lon=add_cyclic_point(ref_seas_data, coord=ref_model_timeser_lon)
        exp_seas_data, exp_model_timeser_lon=add_cyclic_point(exp_seas_data, coord=exp_model_timeser_lon)
    
        #%%create a figure (Robinson-projection) with coastlines
        #fig, ax = plt.subplots(1, 1, dpi=300, subplot_kw={"projection": ccrs.Orthographic(central_longitude=-90, central_latitude=70)}) #old
        fig, ax = plt.subplots(1, 1, dpi=300, subplot_kw={"projection": ccrs.PlateCarree()})
        ax.coastlines(linewidth=0.33,color='gray')
        ax.set_extent((-130,-60,30,80)) # LS: see North america
        ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=False,
                  linewidth=0.5, color='gray', alpha=0.5, linestyle='-')
    
        #%%define color axis for the plot (must be adjusted to the value range of the data))
        level_min=-30 #LS adjust
        level_max=30 #LS adjust
        stepsize=2
        levels=np.arange(level_min-stepsize/2,level_max+stepsize/2+stepsize,stepsize)
        cmap=colors.LinearSegmentedColormap.from_list('mycmap', 
#            ['blue', 'cornflowerblue', 'aqua', 'lightcyan', 'white', 'yellow', 'orange', 'red', 'maroon'], #LS adjust
            ['maroon', 'red', 'orange', 'yellow', 'white', 'lightcyan', 'aqua', 'cornflowerblue', 'blue'], #LS adjust
            (level_max-level_min)/stepsize+1) #+1 adds one level for the center
    
        #%%plot the data anomaly as filled contours
        cf=ax.contourf(ref_model_lon, ref_model_lat, anomaly, transform=ccrs.PlateCarree(),
    	   cmap=cmap, levels=levels, 
    	   vmin=level_min-.5, vmax=level_max+.5,
    	   extend='both')
        
        #%%compute significance matrix for the anomaly between reference and experiment
        sig_mask=compute_significance(exp_seas_data,ref_seas_data,ref_model_lon,ref_model_lat,mask_cutoff=0.95,verbose=False)
        mask=np.ma.masked_not_equal(sig_mask, 1)
        
        #%%overlay hatches that mark regions of insignificant difference between simulation and reference
        cfh=ax.contourf(ref_model_lon,ref_model_lat,mask,transform=ccrs.PlateCarree(),colors='None',extend='both', hatches=['........',None])

        #%%plot a colorbar
        if plot_cbar:
            cax=plt.colorbar(cf, ticks=np.arange(level_min+1,level_max+1-1,2), drawedges=True, orientation='horizontal',
                       label=r"rainfall anomaly [mm/month]",pad=0.10,shrink=1.25,aspect=40) #LS adjust to variable name and unit
            cax.ax.xaxis.set_ticklabels([label.get_text().replace("−", "-") for label in cax.ax.xaxis.get_ticklabels()])
            cax.ax.tick_params(labelsize=9)
            cax.ax.xaxis.set_label_position('top')
        
        #%%export the figure to PNG and PS
        if not plot_cbar:
            outfile=outfile.replace(".png","_no_cbar.png")
        plt.savefig(outfile,dpi=600,bbox_inches='tight')
        print("created file " + outfile)
        plt.close(fig)
        
        #clean up
        cmd="rm reference* *mean.nc *aprt_mm_*"
        print("executing command: " + cmd)
        os.system(cmd)
        cmd="mv *png ../Figures/."
        print("executing command: " + cmd)
        os.system(cmd)
        
```
